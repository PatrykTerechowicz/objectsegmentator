```python
import model
import torch
import matplotlib.pyplot as plt
```

```python
shape = (320, 320)
est_mask = torch.zeros(shape)
gt_mask = torch.zeros(shape)

est_mask[100:200, 100:200] = 1.0
gt_mask[120:230, 90:210] = 1.0
```

```python
plt.imshow(est_mask)
plt.title("ESTIMATED")
plt.show()
plt.imshow(gt_mask)
plt.title("GT")
plt.show()
intersection = est_mask*gt_mask
union = est_mask+gt_mask-intersection
plt.imshow(intersection)
plt.title("INTERSECTION")
plt.show()
plt.imshow(union)
plt.title("UNION")
plt.show()
```

```python
torch.sum(intersection)/torch.sum(union)
```

```python
def calc_iou(out_mask, true_mask):
    nominator = torch.sum(out_mask * true_mask)
    denominator = torch.sum(out_mask) + torch.sum(true_mask) - nominator
    return nominator/denominator
```

```python
calc_iou(est_mask, gt_mask)
```
